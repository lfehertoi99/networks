```python
import scipy as sp
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
def contribution(G,node):
    r=0
    paths1=0
    links=G.degree(node)
    for j in G.nodes():
        for k in G.nodes():
            if j<k and j!=node and k!=node:
                if nx.has_path(G,j,k):
                    paths1+=1
                    r+=nx.shortest_path_length(G, source=j, target=k) 
    r/=paths1 #r is average path length in network with node included. We do not count paths where the node is a source/target.
    print(G.number_of_edges())
    G_new=G.copy()
    G_new.remove_node(node)
    
    s=0
    paths2=0
    for j in G_new.nodes():
        for k in G_new.nodes():
            if j<k:
                if nx.has_path(G_new,j,k):
                    paths2+=1
                    s+=nx.shortest_path_length(G_new, source=j, target=k)
    s/=paths2 #s is average path length in network with node removed.
    print(G_new.number_of_edges())
    if links!=0:
        contribution= (s-r)/links #reduction in average path length
        return contribution
    else:
        return None
```

```python
def newSocialNetwork(m0,mr,ms,Nmax,p,initialcoop):
    
    G = nx.erdos_renyi_graph(m0,p)
    N=m0
    #start with random network
    for i in range (Nmax-m0): #grow to desired size
        G.add_node(m0+i) 
        for node in G.nodes():
            x=np.random.rand()
            if x<=(mr/N): #probability of selecting each individual node as an initial contact
                if node != m0+i:
                    G.add_edge(node,m0+i)
                    for neighbor in G.neighbors(node):
                        y=np.random.rand()
                        if y<=(ms/N): #probability of selecting the neighbour of an initial contact as a secondary contact
                            G.add_edge(neighbor,m0+i)
        N=G.number_of_nodes()
    return G
```

```python
G=newSocialNetwork(10,1.5,0.02,150,0.01,0.15)
```

```python
for node in G.nodes():
    print("contribution of node",node,":",contribution(G,node))
```
