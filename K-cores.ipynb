```python
#G=nx.erdos_renyi_graph(500,0.02)
def k_core_decomposition(G):
    attr=dict()
    for node in G.nodes():
        nodeDict=dict()
        nodeDict["k-shell index"]=0
        attr[node]=nodeDict

    nx.set_node_attributes(G,attr)

    G_copy=G.copy()

    degrees=[]
    for (i,j) in G_copy.degree():
        degrees.append((i,j))
        
    while degrees!=[]:
        minimal_degree=min(d for (n,d) in degrees)
        #print(minimal_degree)
        for (n,d) in degrees:
            if d==minimal_degree:
                #print("node:",n)
                nodeDict=dict()
                nodeDict["k-shell index"]=d
                attr[n]=nodeDict
                G_copy.remove_node(n)

        degrees=[]
        for (i,j) in G_copy.degree():
            degrees.append((i,j))

    nx.set_node_attributes(G,attr)
    
    return G
```
